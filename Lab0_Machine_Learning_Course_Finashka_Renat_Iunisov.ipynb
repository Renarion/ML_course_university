```python
# 2
def arithmetic_diag_matrix(N: int, first: float, diff: float):
    diag_elements = np.array([first + i * diff for i in range(N)])
    matrix = np.diag(diag_elements)
    return matrix

print(arithmetic_diag_matrix(3, 2.0, 3.0))
```

```python
#5
matrix1 = np.array([[5 if (i + j) % 2 == 0 else 0 for j in range(5)] for i in range(5)])
print(matrix1)

matrix2 = np.array([[5 if i <= j else 0 for j in range(5)] for i in range(5)])
print(matrix2)

det1 = np.linalg.det(matrix1)
det2 = np.linalg.det(matrix2)
print("Детерминант матрицы 1:", det1)
print("Детерминант матрицы 2:", det2)

if det1 != 0:
    inv1 = np.linalg.inv(matrix1)
    print(inv1)
else:
    print("Детерминант равен 0 у первой")

if det2 != 0:
    inv2 = np.linalg.inv(matrix2)
    print(inv2)
else:
    print("Детерминант равен 0 у второй")
```

```python
#20
import scipy.stats as st

price = df["Цена"].dropna()
density = st.gaussian_kde(price)
xs = np.linspace(price.min(), price.max(), 200)
density_vals = density(xs)

plt.figure(figsize=(8,6))
plt.plot(xs, density_vals)
plt.xlabel("Цена")
plt.ylabel("Плотность")
plt.title("Ядерная оценка плотности цены")
plt.show()

price_data = df["Цена"].dropna()
floor_data = df["Этаж"].dropna()

data = np.vstack([price_data, floor_data])
kde = st.gaussian_kde(data)

xmin, xmax = price_data.min(), price_data.max()
ymin, ymax = floor_data.min(), floor_data.max()
X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
positions = np.vstack([X.ravel(), Y.ravel()])
Z = np.reshape(kde(positions).T, X.shape)

plt.figure(figsize=(8,6))
plt.contourf(X, Y, Z, cmap='viridis')
plt.xlabel("Цена")
plt.ylabel("Этаж")
plt.title("Двумерная ядерная оценка плотности (Цена и Этаж)")
plt.colorbar()
plt.show()
```
